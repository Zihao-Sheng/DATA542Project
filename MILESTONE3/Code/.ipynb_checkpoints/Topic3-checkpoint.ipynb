## Step1: Analysis Based on Efficiency and Popularity

To explore overall patterns in repository behavior, we applied a clustering analysis again using the two composite metrics constructed earlier—Efficiency and Popularity. 

We first loaded the cleaned dataset (repo_activity.csv) and selected the key variables needed for clustering, including repository identifiers, activity scores, and several metadata fields. The Efficiency and Popularity columns were extracted as the feature matrix and standardized using StandardScaler, which ensures that both metrics contribute equally to the distance calculations in KMeans. 

We then ran a KMeans model with five clusters, assigning each repository to a group based on its standardized feature values. 

Finally, we visualized the clustering results by plotting Efficiency against Popularity, coloring each repository by its assigned cluster to reveal the overall distribution and structure of repository performance.



```python
data=pd.read_csv('repo_activity.csv')
```

```python
data.head()
```

```python
data=data[['id','Efficiency','Popularity','stars','forks','pr_count','dominate_agent']]
```

```python
data
```

```python
X = data[["Efficiency", "Popularity"]].copy()
# data normalization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# setting cluster count
k = 5

# Kmeans
kmeans = KMeans(
    n_clusters=k,
    n_init=10,
    random_state=42
)

labels = kmeans.fit_predict(X_scaled)

# record cluster value
data = data.copy()
data["cluster"] = labels

print("KMeans finished.")
data.head()
```

## Step3:Building a Weighted Repository Network Based on Agent and User Co-Activity
To analyze how repositories are connected through shared development activity, we constructed a weighted network graph using only the high-value repositories identified in the previous step.

We began by filtering the full PR dataset to include only pull requests from these selected repositories, retaining key fields such as repo_id, agent, user_id, and merged_at. Each repository was then added to an undirected graph as a node, and we assigned Efficiency and Popularity scores as node attributes to support later visualization and analysis. 

Edges between repositories were created based on co-activity: if the same agent contributed to multiple repositories, we added edges between those repos and incremented their weights by **1** for each shared agent. Similarly, if the same user ID appeared across multiple repositories—a stronger signal of shared ownership or collaboration—we increased the edge weights by an additional **3** for each such connection. 

This process produced a weighted network that captures meaningful relationships among top-performing repositories, reflecting both agent-level and user-level interaction patterns.


```python
all_pr_df = pd.read_parquet("hf://datasets/hao-li/AIDev/all_pull_request.parquet")
```

```python
all_pr_df.head()
```

```python
#extract neccesary information from data
new = all_pr_df[all_pr_df['repo_id'].isin(temp['id'])].copy()
new = new[['repo_id','agent','user_id','merged_at']]
```

```python
new.head()
```

```python
#Calculating connections between repos
temp['id'] = temp['id'].astype(int)
new['repo_id'] = new['repo_id'].astype(int)

# --- Step 1: Adding all repos as nodes ---
G = nx.Graph()
G.add_nodes_from(temp['id'].unique())

# --- Step 1.5: Giving nodes attributes ---
node_attr = (
    temp
    .set_index('id')[['Efficiency', 'Popularity']]
    .to_dict('index')
)
nx.set_node_attributes(G, node_attr)

# --- Step 2: adding weight for the connection by common agent ---
for agent, df_agent in new.groupby('agent'):
    repos = df_agent['repo_id'].unique()
    for a, b in combinations(sorted(repos), 2):
        if G.has_edge(a, b):
            G[a][b]['weight'] += 1
        else:
            G.add_edge(a, b, weight=1)

# --- Step 3: adding weight for the connection by common user ---
for user, df_user in new.groupby('user_id'):
    repos = df_user['repo_id'].unique()
    for a, b in combinations(sorted(repos), 2):
        if G.has_edge(a, b):
            G[a][b]['weight'] += 3
        else:
            G.add_edge(a, b, weight=3)
```

## Step4: Visualizing the Repository Interaction Network in 3D

To better understand the structural relationships among the selected high-value repositories, we created an interactive 3D graph visualization based on the weighted network constructed earlier. 

Using the NetworkX graph G, which contains repo nodes enriched with Efficiency and Popularity attributes and edges weighted by shared agent and user contributions, we first generated a two-dimensional spring layout and extended it into three dimensions by adding random z-coordinates. Edge weighting was incorporated into the visualization by separating strong edges (weight ≥ 2) from weaker connections, rendering high-weight edges individually with color and thickness proportional to their normalized weight, while sampling weaker edges and displaying them in a unified light-gray layer to reduce visual clutter. 

Node appearance was mapped to repository attributes: Efficiency was encoded using a continuous color scale, and Popularity was reflected in node size. Additional colorbars were included to interpret both node efficiency and edge strength. 

The final 3D plot, built with Plotly, allows for rotation and interactive inspection, providing an intuitive view of how repositories cluster and connect through shared agent and user activity.

```python
%matplotlib widget
import matplotlib.pyplot as plt
```

```python
import networkx as nx
import numpy as np
from collections import defaultdict
import random
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap


# ---------- 0. matplotlib colormap -> plotly colorscale ----------
def mpl_to_plotly(cmap, n=256):
    xs = np.linspace(0, 1, n)
    colors = cmap(xs)
    scale = []
    for x, (r, g, b, a) in zip(xs, colors):
        scale.append([float(x),
                      f"rgb({int(r*255)},{int(g*255)},{int(b*255)})"])
    return scale

# colormap：turbo 
base = plt.cm.turbo
colors_turbo = base(np.linspace(0.1, 1.00, 256))
turbo_green_red = LinearSegmentedColormap.from_list(
    "turbo_green_red", colors_turbo
)
node_colorscale = mpl_to_plotly(turbo_green_red)

# edge inferno（
edge_cmap = plt.cm.inferno
edge_colorscale = mpl_to_plotly(edge_cmap)

# ---------- 1. setting 3d point position ----------
pos2d = nx.spring_layout(G, seed=42, k=1.0, iterations=75)

pos3d = {}
for n, (x, y) in pos2d.items():
    z = np.random.uniform(-1, 1)  # z 
    pos3d[n] = (x, y, z)

# ---------- 2. setting edge for display, the edge number is too big, for edge with weight 5, we are randomly displaying 5 for each node ----------
edges_data = list(G.edges(data=True))

strong_edges = []             # weight >= 2
weak_incident = defaultdict(list)
edge_attr = {}

for u, v, d in edges_data:
    w = d.get("weight", 1)
    key = tuple(sorted((u, v)))
    edge_attr[key] = (u, v, d)

    if w >= 2:
        strong_edges.append((u, v, d))
    elif w == 1:
        weak_incident[u].append(key)
        weak_incident[v].append(key)

chosen_weak_keys = set()
max_weak_per_node = 5
random.seed(42)

for node, edge_keys in weak_incident.items():
    if len(edge_keys) <= max_weak_per_node:
        chosen = edge_keys
    else:
        chosen = random.sample(edge_keys, max_weak_per_node)
    chosen_weak_keys.update(chosen)

weak_edges_selected = [edge_attr[key] for key in chosen_weak_keys]

# ---------- 3. normalize edge weight for coloring ----------
all_w = np.array([d.get("weight", 1) for _, _, d in edges_data])
w_min, w_max = all_w.min(), all_w.max()

def norm_w(w):
    return (w - w_min) / (w_max - w_min + 1e-9)

# ---------- 4. weak edge trace（ 3D trace） ----------
edge_x_weak, edge_y_weak, edge_z_weak = [], [], []

for u, v, d in weak_edges_selected:
    x0, y0, z0 = pos3d[u]
    x1, y1, z1 = pos3d[v]
    edge_x_weak += [x0, x1, None]
    edge_y_weak += [y0, y1, None]
    edge_z_weak += [z0, z1, None]

edge_trace_weak = go.Scatter3d(
    x=edge_x_weak,
    y=edge_y_weak,
    z=edge_z_weak,
    mode='lines',
    line=dict(width=1, color='rgba(200,200,200,0.25)'),
    hoverinfo='none',
    showlegend=False
)

# ---------- 5. strong edge trace：3D  ----------
strong_edge_traces = []

for u, v, d in strong_edges:
    x0, y0, z0 = pos3d[u]
    x1, y1, z1 = pos3d[v]
    w = d.get("weight", 1)
    nw = norm_w(w)

    rgba = edge_cmap(nw)
    color_str = f"rgba({int(rgba[0]*255)},{int(rgba[1]*255)},{int(rgba[2]*255)},{rgba[3]:.2f})"

    trace = go.Scatter3d(
        x=[x0, x1],
        y=[y0, y1],
        z=[z0, z1],
        mode='lines',
        line=dict(
            width=2.0 + 4 * nw,
            color=color_str
        ),
        hoverinfo='none',
        showlegend=False
    )
    strong_edge_traces.append(trace)

# ---------- 6. node  ----------
nodes = list(G.nodes())
x_nodes = [pos3d[n][0] for n in nodes]
y_nodes = [pos3d[n][1] for n in nodes]
z_nodes = [pos3d[n][2] for n in nodes]

eff = np.array([G.nodes[n].get("Efficiency", 0) for n in nodes])
e_min, e_max = eff.min(), eff.max()
norm_eff = (eff - e_min) / (e_max - e_min + 1e-9)
gamma = 0.8
norm_eff_adj = norm_eff ** gamma   # 0~1，

pop = np.array([G.nodes[n].get("Popularity", 0) for n in nodes])
pop_scaled = 5 + 25 * (pop - pop.min()) / (pop.max() - pop.min() + 1e-9)

hover_text = [
    f"id: {n}<br>Efficiency: {G.nodes[n].get('Efficiency', 0):.2f}"
    f"<br>Popularity: {G.nodes[n].get('Popularity', 0):.2f}"
    for n in nodes
]

# ---------- 7. nodes 3D trace ----------
node_trace = go.Scatter3d(
    x=x_nodes,
    y=y_nodes,
    z=z_nodes,
    mode='markers',
    hoverinfo='text',
    text=hover_text,
    marker=dict(
        size=pop_scaled,
        color=norm_eff_adj,           # 0~1
        colorscale=node_colorscale,   # 浅绿 -> 红
        cmin=0, cmax=1,
        line=dict(width=0.5, color='black'),
        colorbar=dict(
            title="Efficiency",
            x=1.08
        )
    ),
    showlegend=False
)

# Edge Weight 
edge_colorbar_dummy = go.Scatter3d(
    x=[None],
    y=[None],
    z=[None],
    mode='markers',
    marker=dict(
        size=0,
        color=[w_min, w_max],
        colorscale=edge_colorscale,
        cmin=w_min,
        cmax=w_max,
        colorbar=dict(
            title="Edge Weight",
            x=1.18
        )
    ),
    showlegend=False,
    hoverinfo='none'
)

# ---------- 8. all trace ----------
data = [edge_trace_weak] + strong_edge_traces + [node_trace, edge_colorbar_dummy]

fig = go.Figure(data=data)

fig.update_layout(
    width=1000,
    height=1600,
    title="Repo Graph 3D (drag to rotate, hover node for details)",
    showlegend=False,
    scene=dict(
        xaxis=dict(showbackground=False, visible=False),
        yaxis=dict(showbackground=False, visible=False),
        zaxis=dict(showbackground=False, visible=False),
        aspectmode='data'
    ),
)

fig.show()
```

## Step 6:Detecting Structural Communities Using the Louvain Algorithm

To uncover cohesive structural groups within the refined repository network, we applied the Louvain community detection algorithm. 

Using edge weights as the optimization criterion, the algorithm partitions the graph into communities that maximize modularity, revealing clusters of repositories that share strong collaborative or behavioral connections. 

After computing the partition, we grouped repositories by their assigned community and inspected the size and composition of each group. 

For visualization, we generated a 2D spring layout and colored nodes according to their community labels, enabling a clear view of how repositories cluster based on shared agent and user activity. This step highlights the underlying structure of the repo network and identifies subgroups that operate more closely together.

```python
import community as community_louvain
partition = community_louvain.best_partition(G, weight='weight')
```

```python
print(partition)
```

```python
from collections import defaultdict

groups = defaultdict(list)
for node, com in partition.items():
    groups[com].append(node)

for com, nodes in groups.items():
    print(f"Community {com}: {len(nodes)} nodes")
    print(nodes[:20])
    print("---")
```

```python
community_df = pd.DataFrame.from_dict(partition, orient="index", columns=["community"])
community_df.index.name = "repo"
```

```python
%matplotlib inline
# 1. Community detection（Louvain）
partition = community_louvain.best_partition(G, weight='weight')

# 2. Coloring nodes
node_colors = [partition[n] for n in G.nodes()]

# 3. Spring layout
pos = nx.spring_layout(G, k=2.5, iterations=100, seed=42)

plt.figure(figsize=(14, 12))

nx.draw(
    G, pos,
    node_color=[partition[n] for n in G.nodes()],
    cmap=plt.cm.tab20,
    node_size=100,
    width=0.3,
    alpha=0.8
)

# drawing labels
labels = {n: n for n in G.nodes()}
label_pos = {n: (x, y+0.02) for n,(x,y) in pos.items()}  # 上移一点

nx.draw_networkx_labels(G, label_pos, labels=labels, font_size=6)
plt.axis("off")
plt.show()
```

# Final Analysis of Contributor-Based Communities in the AI Agent Ecosystem

## 1. Overview
After performing contributor‑based community detection on a cleaned mutual‑contributor graph, the ecosystem separates into **two large communities** and **two small communities**.  
These communities reflect distinct modes of development within the contemporary AI‑enhanced software engineering landscape.  
The following sections summarize the structural, linguistic, and behavioral characteristics of these communities in a consolidated academic format.

---

## 2. Shared Characteristics of the Two Large Communities
Despite addressing different layers of the AI engineering stack, the two major communities share several fundamental traits that define them as the “core ecosystem hubs” of the AI‑agent developer network.

### **2.1 Dominant AI Agent Usage**
In both communities, **OpenAI Codex** functions as the overwhelmingly dominant agent used by contributors during development work.  
Although Devin and Cursor appear in a limited number of repositories, Codex’s prevalence suggests that contributors rely primarily on **AI‑assisted coding workflows**, rather than autonomous or planning‑centric agents.  
This dominance is consistent across languages and repository sizes.

### **2.2 Repository Topology: “Core Repository + Peripheral Satellites”**
Both large communities exhibit a **hub‑and‑spoke topology**:
- A small number of highly starred, publicly visible repositories (e.g., *mlflow/mlflow*, *gofiber/fiber*, *langfuse-docs*) function as the **public-facing cores**.
- These are surrounded by a large number of **low‑star, low‑fork but high‑PR‑activity** repositories representing auxiliary tools, experimental features, integration layers, or personal engineering sandboxes.

This indicates that contributors maintain a **professional, multi-repository workflow**, where a dominant “main project” is supported by numerous smaller codebases.

### **2.3 Multi‑Language Engineering Environment**
Both major communities operate within a **multi‑language, multi‑layered technical ecosystem**.  
Languages commonly include TypeScript, JavaScript, Python, Go, Rust, C#, and occasionally PHP or Ruby.  
This diversity suggests that contributors are engaged in **full-stack AI engineering**, spanning:
- front-end interfaces,
- backend orchestration layers,
- model-serving gateways,
- performance-critical modules,
- and domain-specific utilities.

### **2.4 Open Licensing Culture with Experimental Subspaces**
Licenses across the two major communities are dominated by **MIT** and **Apache-2.0**, but also include several repositories with *None* or *NOASSERTION*.  
This mixture indicates:
1. A **commitment to permissive open-source practices**, and  
2. The presence of **rapid prototyping spaces**, where licensing has not yet stabilized.

---

## 3. Distinct Roles of the Two Large Communities

### **3.1 Community A: Web‑Oriented AI Agent Engineering**
The first large community clusters around repositories involving:
- agent dashboards and control panels,
- LLM gateways,
- front-end interaction systems,
- observability and logging documentation.

It leans heavily toward **TypeScript/JavaScript ecosystems**, indicating an emphasis on UI/UX, web orchestration, and agent-facing user layers.

### **3.2 Community B: ML‑Oriented AI Engineering and Traditional DevOps**
The second large community centers around:
- MLflow and similar ML engineering frameworks,
- general Python-based AI applications,
- high-performance Go infrastructure (e.g., *fiber*),
- small automation or analytic tools.

This community reflects a hybrid between **traditional ML engineers** and **agent-augmented developers**, with Copilot appearing as an additional agent in classical ML workflows.

---

## 4. Characteristics of the Two Small Communities
The two small communities diverge from the structural complexity of the large ones.  
Both are **internally homogeneous** and represent *localized development clusters*, typically tied to a single author or a single organization.

### **4.1 Community C (splanck/*): C-Based Toolchain Cluster**
This community contains several C-language repositories entirely belonging to the *splanck/* namespace, all licensed under BSD‑2-Clause.  
Their extremely high PR counts but zero public stars/forks suggest internal or specialized system-level codebases.  
All rely exclusively on **OpenAI Codex**, demonstrating a strong preference for AI-assisted low-level code generation in a controlled, focused environment.

### **4.2 Community D (adrianwedd/*): Python Agentic Computing Projects**
This community contains several Python repositories under the *adrianwedd/* namespace, each involving “Agentic” terminology or AI research tools.  
These repositories also exhibit very low public visibility but high internal development activity.  
Again, Codex is the sole agent used, pointing to a consistent workflow by a single developer or a tightly coordinated small team.

---
